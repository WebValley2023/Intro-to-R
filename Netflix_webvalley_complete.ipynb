YEAR ADDED

Time series data

check na values

```r
table(is.na(netflix$date_added))
```

Transform data_added column in date format

```r
netflix <- netflix %>% 
  mutate(date_added = mdy(date_added))
```

Create new variables starting from data_added 

Getting each year, month and date separated 

```r
netflix <- netflix %>% 
  mutate(Year = year(date_added)) %>% 
  mutate(Month = month(date_added, label = TRUE)) %>% 
  mutate(Date = day(date_added)) %>% 
  mutate(Day = wday(date_added, label = TRUE))
```

Number of Items added by year 

```r
df_plot_1 <- netflix %>%
  group_by(Year) %>%
  #group_by(Year,type) %>%
  summarise(total = n())

ggplot(df_plot_1) +
  theme_minimal() +
  geom_col(aes(
    x = Year,
    y = total
    #, fill = type))+
    ), fill = "greenyellow")+
  labs(title = "Number of Items added by year") 
```

Number of Items added each day of the week

```r
netflix %>%
  #filter(!is.na(Day)) %>%
  group_by(Day) %>%
  summarise(total = n()) %>%
  ggplot() +
  theme_minimal() +
  geom_col(aes(
    x = Day,
    y = total),
    fill="cornflowerblue")+
  labs(title = "Number of Items added each day of the week") 
```

Let's analyse data by quarters

```r
df_plot_2 <- netflix %>%
  group_by(Year,  m=quarter(date_added)) %>%
  #group_by(Year,type) %>%
  summarise(total = n())%>%
  filter(!is.na(Year))
df_plot_2$YM = str_c(df_plot_2$Year,"-", df_plot_2$m)
```

Line plot of number of item added by year quarter

```r
ggplot(data=df_plot_2, aes(
  x=YM, 
  y=total, 
  group=1
  )) +
  geom_line()+
  geom_point()+
  labs(
      x = "Year Added",
      y = "Count",
      title = "Number of Items added by year quarter"
    )+
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))
```

```r
list <- as.list(netflix$genres)
string <- toString(list)
genres <- str_extract_all(string, "[a-z]+")
list_of_genres <- as.data.frame(genres, col.names = c("genre"))
top_genres <- list_of_genres %>% 
  group_by(genre) %>% 
  summarise(count = n())  %>% 
  arrange(desc(count)) 
```

```r
ggplot(top_genres)+
  geom_col(aes(count, reorder(genre, count)), fill="#089BA3")+
  theme_minimal()+
  labs(title = "The count of vidios by genres")
```

```r
netflix$decade = NA 

for (i in 1:dim(netflix)[1])  {
  netflix$decade[i] = netflix$release_year[i] - netflix$release_year[i] %% 10
}
```

```r
netflix$decades2 <- netflix$release_year - netflix$release_year %% 10
```

```r
netflix %>% 
  filter(!is.na(tmdb_popularity)) %>%
  group_by(decades2) %>%
  summarise(Mean = mean(tmdb_popularity)) %>%
  ggplot()+
  geom_col(aes(decades2, Mean), fill="#089BA3")+
  theme_minimal()+
  labs(title = "")
```

EXTRA for the curious and brave ones

```r
library(wordcloud)
library(tidytext)
desc_words_m <- netflix %>% 
  filter(type == "MOVIE" ) %>% 
    unnest_tokens(word, description) %>%
    anti_join(stop_words)

count_word <- desc_words_m %>%
   count(word, sort = TRUE)
```

```r
wordcloud(words = count_word$word,  
          freq = count_word$n,
          min.freq = 50,  
          max.words = nrow(count_word), 
          random.order = FALSE,  
          rot.per = 0.1,  
          colors = brewer.pal(8, "Dark2")
)
```
